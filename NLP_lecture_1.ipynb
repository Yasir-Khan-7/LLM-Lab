```python
import re
```

```python
chat1 = "codebasics: you ask lot of questions 😠  1235678912, abc@xyz.com"
chat2 = 'codebasics: here it is: (123)-567-8912, abc@xyz.com'
```

```python
pattern = r"\(\d{3}\)-\d{3}-\d{4}| \d{10}"
pattern2 = r"[a-z0-9A-Z_]*@[a-z0-9A-Z]*\.[a-z]*"
```

```python
matches = re.findall(pattern, chat1)
matches += re.findall(pattern, chat2)
matches += re.findall(pattern2, chat1)
print(matches)
```

```python
def pattern_generator(pattern,text):

    matches = re.findall(pattern,text)
    if matches:
        return matches[0]

   
```

```python
def information_extractor(text):

    age = pattern_generator(r"age (\d+)", text)
    born = pattern_generator(r"Born(.*)", text)
    return age, born
```

```python
information_extractor(wikepediatext)
```
